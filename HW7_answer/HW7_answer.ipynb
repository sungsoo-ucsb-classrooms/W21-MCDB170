#### Problem 1 (2 pt)

Write a function that calculates covariance of two 1-dimensional arrays.

1. This function receives two 1-dimensional arrays: `x` and `y`.

2. It calculates the covariance between two arrays using the following formula.

    $cov(x,y) = \frac{1}{N}\sum_{n=0}^{N-1}(x_n-\bar x)(y_n-\bar y)$
    
    where, $x_n$ is the n-th element in the array, $\bar x$ is the average of the array (likewise for $y$), and $N$ is the total number of elements in the array.

3. The result should be returned.


Hint: Use ndarray.mean() and ndarray.size.

The output should look like below:



```python
# Change the following function.

# DO NOT CHANGE THE FUNCTION NAME AND ARGUMENTS!!!

import numpy as np


def get_cov( x, y ):
    return ((x-x.mean())*(y-y.mean())).sum()/x.size
```

```python
# The following code is provided for your test.
# You can modify it as you want, to test your function.

np.random.seed(100)
x = np.random.normal(size=[100])
y = np.random.normal(size=[100])

print(get_cov(x,y))
```

#### Problem 2 (3 pt)

Write a function that returns a covariance matrix.

1. This function receives a 2-dimensional array (i.e., matrix), `x`. The rows of this matrix are individual observations, and the columns represent different variables. We will calculate the covariance between different columns.

2. First, get the dimension of the input array and put it into two variables `m` and `n`. `m` should store the number of rows, and `n` should store the number of columns.

3. Second, make a matrix named `covmat` with zeros and with the size of $n$ x $n$. This will hold the covariance matrix.

4. Then, make a nested loop to select pairs of columns and calculate covariance usng the `get_cov()` function in Problem 1. The result should be stored in the `covmat`.

    $covmat_{i,j} = cov( x[:,i] , x[:,j]  )$
    
    where, $covmat_{i,j}$ is the element of `covmat` matrix at row `i` and column `j`, $cov()$ is the function you developed in Problem 1, $x[:,i]$ represents the i-th column of the matrix `x`.

5. `covmat` should be returned.


The outcome of this problem should look like below:



```python
# Complete the following function.

# DO NOT CHANGE THE FUNCTION NAME AND ARGUMENTS!!!

import numpy as np
def get_covmat( x ):
    m,n = x.shape
    covmat = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            covmat[i,j] = get_cov( x[:,i], x[:,j])
    
    return covmat
    
```

```python
# The following code is provided for your test.
# You can modify it as you want, to test your function.

# Iris data. See https://archive.ics.uci.edu/ml/datasets/Iris
x = np.genfromtxt('iris.data', delimiter=',', dtype='float', usecols=[0,1,2,3])
print(get_covmat(x))
```

#### Problem 3 (2 pt)

Write a function that calculates the correlation matrix.


The procedure is exactly the same as Problem 3, except that you need to use a different formula to calculate correlation matrix:

$corrmat_{i,j} = \displaystyle\frac{cov( x[:,i] ,  x[:,j]  ) }{ std(x[:,i])std(x[:,j])}$

Use `corrmat` instead of `covmat`.


The result should look like below:



```python
# Complete the following function.

# DO NOT CHANGE THE FUNCTION NAME AND ARGUMENTS!!!

import numpy as np
def get_corrmat( x ):
    m,n = x.shape
    corrmat = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            corrmat[i,j] = get_cov( x[:,i], x[:,j]) / x[:,i].std() / x[:,j].std()
    
    return corrmat
    
```

```python
# The following code is provided for your test.
# You can modify it as you want, to test your function.

x = np.genfromtxt('iris.data', delimiter=',', dtype='float', usecols=[0,1,2,3])
print(get_corrmat(x))
```
