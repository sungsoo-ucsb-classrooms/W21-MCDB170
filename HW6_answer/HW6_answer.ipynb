#### Problem 4 (2 pt)

Write a function that calculates L1 deviation of data.

1. This function receives an array.

2. It calculates the L1 deviation (also called an absolute deviation) of the array. Use the following formula.

    $D = \frac{1}{N}\sum_{n=0}^{N-1}|x_n-\bar x|$

    where, $x_n$ is the n-th element in the array, $\bar x$ is the average of the array, and $N$ is the total number of data.
    
    In NumPy, the absolute values of an array `k` can be obtained by using, `np.absolute(k)`.
    
    Hint: Use a combinations of `.sum()`, `.size`, and `np.absolute()`.

3. $D$ should be returned.

The result should look like below:



```python
# Complete the following function.

# DO NOT CHANGE THE FUNCTION NAME AND ARGUMENTS!!!

import numpy as np
def get_deviation( x ):
    x_bar = x.sum()/x.size
    D = np.absolute((x-x_bar)).sum()/x.size
    return D
```

```python
data = np.arange(100)
D = get_deviation(data)
print(D)

data = np.array([5,3,7,9,16,3,9,8,21])
D = get_deviation(data)
print(D)
```

#### Problem 5 (3 pt)

Write a function that returns the coefficients of $(x+y)^n$.

1. This function receives a number $n$.

2. It makes an array of $n+1$ elements. r-th element of this array should be 

    $_n C_r = \frac{n!}{r!(n-r)!}$

    where, $r$ is an integer and $0 \le r \le n$.
    
    Factorial can be computed using a function, `factorial()` from `scipy.special` package. This is already imported in the skeleton code. To obtain element-wise factorial of an array `a`, simply call `factorial(a)` and it will return an array of factorial of each element in the array `a`.
    
3. Retuen the array containing the coefficients.
    
    
cf) Coefficients of $(x+y)^n$ can be obtained with the Pascal's triangl. See [here](https://en.wikipedia.org/wiki/Pascal%27s_triangle). 

The result should look like below:
 


```python
# Complete the following function.

# DO NOT CHANGE THE FUNCTION NAME AND ARGUMENTS!!!

import numpy as np
from scipy.special import factorial

def coefficients( n ):
    num = np.ones(n+1)*n
    den = np.arange(n+1)
    
    return factorial(num)/factorial(den)/factorial(n-den)
    
```

```python
for n in range(10):
    print(coefficients(n))
```
